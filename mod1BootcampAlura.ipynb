```
import pandas as pd
```

```
#Neste caso, usa-se o enconding ISO para representar o padrão UTF-8, skiprows = 3
#para ignorar as 3 primeiras linhas, skipfooter para ignorar as ultimas 12
#,usa-se o separador como ";" em vez de ","
#, usa-se thousands = "." para mostrar o separador de milhares
#, musa-se decimal = "," para mostrar o separador de decimal do arquivo como ","


dados = pd.read_csv("/content/A151346189_28_143_208.csv", encoding="ISO-8859-1",
                    skiprows = 3, sep = ";", skipfooter = 12,
                    thousands=".", decimal=",")

dados.head()  #primeiras 5 linhas
```

```
dados.tail()  #ultimas 5 linhas
```

```
dados.info()
```

```
#media
dados.mean()
```

```
#Troca o formado de apresentacao cientifico para com 2 casas decimais depois do "."
pd.options.display.float_format= "{:.2f}".format
```

```
dados.mean()
```

```
#Acessar os dados de agosto de 2008 (que no caso é uma série)
dados["2008/Ago"]
```

```
dados["2008/Ago"].mean()
```

```
#Gráfico de gasto por unidade federativa
dados.plot(x = "Unidade da Federação",y = "2008/Ago")
```

```
#Porém, o gráfico de linhas não é adequado para essa situação
#Para mudar para gráfico de barra, kind="bar"
#usa-se figsize para alterar o tamanho da figura
dados.plot(x = "Unidade da Federação",y = "2008/Ago", kind = "bar", figsize = (10,7))
```

```
#Vamos importar outra biblioteca gráfica para poder utilizar melhor dos argumentos do gráfico
import matplotlib.pyplot as plt

#Biblioteca para mudar o formato dos eixos
import matplotlib.ticker as ticker

#ax = axis, ou seja, eixo X
ax = dados.plot(x = "Unidade da Federação",y = "2020/Jul", kind = "bar", figsize = (10,7))
#usamos um set para o formato ser um formato float com o separador de milhares como ","
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

#Adicionando um título
plt.title("Valor total para o mês de Agosto em processamento segundo Unidade da Federação")

#Mostrar o gráfico
plt.show()
```

# AULA 02 MOD 01

```
import pandas as pd
```

```
dados = pd.read_csv("/content/A160324189_28_143_208.csv", encoding="ISO-8859-1",
                    skiprows = 3, sep =";", skipfooter = 12,
                    thousands = ".", decimal = ",")
dados.head()
```

```
#Eh necessario passar dois colchetes ao passar mais de uma coluna
dados[["2008/Ago", "2008/Set"]].head()
```

```
#A media ignora os valores vazios que estao antes de 2007
#pois ela ignora os valorez vazios, no caso 2007 começa a ter todos os valores
dados.mean()
```

```
#Ao fazer as colunas usaveis excluimos a unidade da federacao.
#Precisamos entao adiciona-la dnv, porém, precisamos transforma-la em lista
#antes para que possamos inseri-la, pois o index eh uma serie imutavel
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```
#Nota-se que o index está como um número (0,1,2...etc) e não como a unidade
#da federação, então usa-se set_index("Unidade da Federação")
usaveis = usaveis.set_index("Unidade da Federação")
usaveis
```

```
#Para ver uma coluna só
usaveis["2019/Ago"].head()
```

```
#Para localizar uma linha cujo indice é Acre
#O erro no arquivo nos impede de ver, mas funcionaria
usaveis.loc["12 Acre"]
```

```
#exemplo transposição  (as linhas viram colunas e vice versa)
usaveis.T.head()
```

```
#Plotar usaveis em um tamanho 12 por 6 com uma transposição T (as colunas viram
#os estados e as linhas os valores)
usaveis.T.plot(figsize=(10,6))
```

```
#Precisamos remover a linha total para poder plotar um gráfico melhor
#"Procure no eixo 1 (eixo da coluna) a linha total"
usaveis = usaveis.drop("Total",axis = 1)
usaveis.head
```

```
usaveis.T.plot(figsize=(10,6))
#adicionando quadriculado no gráfico
plt.grid(True)
#posicionando a legenda para fora do gráfico
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
```
