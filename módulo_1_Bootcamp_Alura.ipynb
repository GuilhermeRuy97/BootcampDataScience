# AULA 02 MOD 01

```
import pandas as pd
```

```
dados = pd.read_csv("/content/A160324189_28_143_208.csv", encoding="ISO-8859-1",
                    skiprows = 3, sep =";", skipfooter = 12,
                    thousands = ".", decimal = ",")
dados.head()
```

```
#Eh necessario passar dois colchetes ao passar mais de uma coluna
dados[["2008/Ago", "2008/Set"]].head()
```

```
#A media ignora os valores vazios que estao antes de 2007
#pois ela ignora os valorez vazios, no caso 2007 começa a ter todos os valores
dados.mean()
```

```
#Ao fazer as colunas usaveis excluimos a unidade da federacao.
#Precisamos entao adiciona-la dnv, porém, precisamos transforma-la em lista
#antes para que possamos inseri-la, pois o index eh uma serie imutavel
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```
#Nota-se que o index está como um número (0,1,2...etc) e não como a unidade
#da federação, então usa-se set_index("Unidade da Federação")
usaveis = usaveis.set_index("Unidade da Federação")
usaveis
```

```
#Para ver uma coluna só
usaveis["2019/Ago"].head()
```

```
#Para localizar uma linha cujo indice é Acre
#O erro no arquivo nos impede de ver, mas funcionaria
usaveis.loc["12 Acre"]
```

```
#exemplo transposição  (as linhas viram colunas e vice versa)
usaveis.T.head()
```

```
#Plotar usaveis em um tamanho 12 por 6 com uma transposição T (as colunas viram
#os estados e as linhas os valores)
usaveis.T.plot(figsize=(10,6))
```

```
#Precisamos remover a linha total para poder plotar um gráfico melhor
#"Procure no eixo 1 (eixo da coluna) a linha total"
usaveis = usaveis.drop("Total",axis = 1)
usaveis.head
```

```
usaveis.T.plot(figsize=(10,6))
#adicionando quadriculado no gráfico
plt.grid(True)
#posicionando a legenda para fora do gráfico
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
```

```
usaveis[0:5].T.plot(figsize=(10,6))
```

```
#criar uma amostra aleatoria (no caso com n = 7)
#cada vez que se roda o código,  um resultado diferente é mostrado
usaveis.sample(n = 7)
```

```
#criando numeros aleatorios usando uma seed com numpy
import numpy as np
np.random.seed(687942)

usaveis.sample(n  = 7)
```

```
#Fazer a soma de uma linha
#Lembre-se que a função sum sem o parametro somaria a coluna e não a linha
usaveis.sum(axis=1)
```

```
#Criar uma coluna com o total por estado
usaveis["Total"] = usaveis.sum(axis=1)
usaveis.head()
```

```
#Ordenando pelo total utilizando a coluna total 
#em ordem decrescente (ascending = False, já que por padrão é True)
ordenado_por_total = usaveis.sort_values(by = "Total", ascending = False)

#Já que temos a ordenação, podemos excluir a coluna total agora
ordenado_por_total = ordenado_por_total.drop("Total", axis = 1)
ordenado_por_total.head()
```

```
ordenado_por_total.head(5).T.plot(figsize=(10,6))
```

```
#Porém, percebe-se uma anormalidade nos dados de janeiro de 2008m, provavelmente
#dados viciados (adicionados todos em uma data)
#Vamos cortá-los então, pegando os dados da sexta coluna para frente

colunas_interessadas = ordenado_por_total.columns[6:]
ordenado_por_total = ordenado_por_total[colunas_interessadas]
```

```
ordenado_por_total.head(5).T.plot(figsize=(10,6))
```

```
#Mudando as informações da coluna para deixar em milhões
ordenado_por_total = ordenado_por_total / 1000000
ordenado_por_total.head(5).T.plot(figsize=(10,6))
```

```
#Começando a analisar valores para os 3 primeiros que mais gastaram a partir de 2012 até mai/2020
#No caso 2012 se da a partir da linha 47 e mai/2020 se da a partir da linha 150
meses_filtrados = ordenado_por_total.columns[47:150]
ordenado_por_total[meses_filtrados].head(3).T.plot(figsize=(10,6))

#Porém, note que o gráfico está distorcido já que são paulo mostra um resultado
# 4 vezes maior de gasto. Esté é o perigo de mudar os campos do gráfico (criar ilusões)
```

```
#Fazendo o mesmo gráfico anterior mas não distorcendo o campo visual (para não
#distorcer a realidade como no gráfico anterior)
#Faremos isso com o y limite (usando de 0 reais a 350 milhoes)
meses_filtrados = ordenado_por_total.columns[47:150]
ordenado_por_total[meses_filtrados].head(3).T.plot(figsize=(10,6))
plt.ylim(00, 350)
```
