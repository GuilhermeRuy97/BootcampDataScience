## A partir daqui utilizaremos função para refatorar o código acima


```
#Criando função para refatorar código

def insere_gastos_e_gasto_por_habitante_para(todos_os_gastos: pd.DataFrame, dados: pd.DataFrame, mes: str):
  gastos_do_mes = todos_os_gastos[mes]
  dados[f"gastos_{mes}"] = gastos_do_mes * 1000000
  dados[f"gasto_por_habitante_{mes}"] = dados[f"gastos_{mes}"] / dados["populacao"]
```

```
#Tirando os 3 primeiros caracteres para todos e não somente para os mais recentes

ordenado_por_total.index = ordenado_por_total.index.str[3:]
```

```
#trabalhando com data e hora no python
#Existe uma função na biblioteca datetime chamada de datetime

import datetime

datetime.datetime(2020, 11, 3)  #ano/mes/dia
```

```
#importando o dia atual

from datetime import date

date.today()
```

```
#Existe uma função que podemos aplicar a series onde pegamos o índice e aplicamos uma função

#Criando um dicionário para o meses
meses= {
    "Jan": 1,
    "Fev": 2,
    "Mar": 3,
    "Abr": 4,
    "Mai": 5,
    "Jun": 6,
    "Jul": 7,
    "Ago": 8,
    "Set": 9,
    "Out": 10,
    "Nov" : 11,
    "Dez" : 12
}

#função que transforma o mes em dia

def para_dia(ano_mes: str):
  ano: int = int(ano_mes[:4])    #caracter do ano são os ultimos 4 dígitos
  mes: str = ano_mes[5:]         #caracter do mes é a partir do quinto caracter
  mes_numero: int = meses[mes]
  return date(ano, mes_numero, 1)

#mapeando os valores para cada valor
mensal.index = mensal.index.map(para_dia)
mensal.head()

#Repare que o índice agora está como data e não mais uma string
```

```
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=mensal, x=mensal.index, y="São Paulo")
plt.xticks(rotation=30)
plt.ylim(0, 350)
plt.grid(linestyle="--")
plt.show()

#repare como ele automaticamente coloca os anos e marca nos janeiros agora
```

```
#Vendo os tipos das colunas

display(mensal.info())

#repare como o indice é uma data de 2008-02-01 a 2020-07-01
#pois, como agora temos uma data ao invés de uma string, o próprio compilador
#consegue fazer um melhor "locator" automático

display(mensal.index[0])
```

```
#Vamos agora plotar o gráfico para todos os estados
#Primeiro resetamos o indice
#Com o .met() , "derretemos" os estados separadamente e dizemos como parâmetro quem não queremos derreter
#Falamos então que queremos manter a coluna de data, que se chama index
#Forçamos uma ordem passando os estados na ordem em que aparecem

mensal_aberto = mensal.reset_index().melt(id_vars = ["index"], value_vars = mensal.columns)
mensal_aberto.columns = ["mes", "uf", "gasto"]    #trocando os nomes das colunas
mensal_aberto
```

```
plt.figure(figsize=(10,6))
ax = sns.lineplot(data = mensal_aberto, x = "mes", y = "gasto")
plt.xticks(rotation = 30)
plt.ylim(0, 350)
plt.grid(linestyle = "--")
plt.show()

#A linha azul é a média de valores e a faixa é o intervalo de confiança
```

```
#agrupando por uf utilizando o hue

plt.figure(figsize=(10,6))
ax = sns.lineplot(data = mensal_aberto, x = "mes", y = "gasto", hue = "uf")
plt.xticks(rotation = 30)
plt.ylim(0, 350)
plt.grid(linestyle = "--")
plt.show()
```
