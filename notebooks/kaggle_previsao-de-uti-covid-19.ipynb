```python
#Buscando quais são as colunas númericas para utilizar depois;
#O x e y precisam ter o mesmo tamanho.

x_columns = dados.describe().columns
dados_limpos = dados.dropna()
```

```python
#A entrada é "tudo menos o ICU" e a saida é o ICU

y = dados_limpos["ICU"]
x = dados_limpos[x_columns].drop(["ICU"], axis = 1)
```

```python
#Trabalhando com o scikit learn

from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression()
modelo.fit(x, y)
```

```python
#Vendo a porcentagem de acerto

accuracy = sum(modelo.predict(x) == y) / len(y) * 100
accuracy

#Ou seja, 91%
```

Porém, não é um modelo muito inteligente, pois damos os valores x e y e
testamos com os mesmos x e y. Ele então, pode ter "decorado" somente para aqueles
dados e não será capaz de prever algo. <br><br>

Também é necessário notar que o modelo pode chutar a classificação mais comum,
por exemplo, chutar que uma pessoa não tem HIV pois a maioria das pessoas não
tem HIV (caso ele chutasse não para todos, acertaria em 99.6% das vezes).

```python
#Brincando com o DummyClassifier, que por padrão, chuta de acordo com a proporção;
#Note como ele acerta 47% das vezes

from sklearn.dummy import DummyClassifier

modelo = DummyClassifier()
modelo.fit(x, y)
sum(modelo.predict(x) == y) / len(y) * 100
```

```python
#Podemos medir a acurácia de um modelo com o accuracy_score

from sklearn.metrics import accuracy_score

accuracy_score(y, modelo.predict(x))
```

```python
#Precisamos separar os dados de teste e treino

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
print(len(x_train))
print(len(y_train))

#No caso, ele treinará com 408 pacientes

print(len(x_test))
print(len(y_test))

#Estes são os pacientes remanescentes usados para teste
```

```python
modelo = LogisticRegression()
modelo.fit(x_train, y_train)
y_prediction = modelo.predict(x_test)
accuracy_score(y_test, y_prediction)

#O resultado é de 89%, este resultado muda conforme a execução (cada vez que se
#faz um run all o valor muda);
#Isso pode acontecer pelo numero de dados para treino e testes estarem desbalanceados.
```

```python
#Podemos separar de acordo com as classes, por exemplo, se uma classe aparece 51%
#das vezes nos dados de treino, ela aparecerá 51% das vezes nos dados de teste;
#Damos a isto o nome de estratificar stratify = true (y).

x_train, x_test, y_train, y_test = train_test_split(x,y, stratify = y)

modelo = LogisticRegression()
modelo.fit(x_train, y_train)
y_prediction = modelo.predict(x_test)
accuracy_score(y_test, y_prediction)
```

```python
#Começando do 0 novamente

dados = pd.read_excel("/kaggle/input/covid19/Kaggle_Sirio_Libanes_ICU_Prediction.xlsx")
dados.head()
```

```python
features_continuas = dados.iloc[:, 13:-2].columns
features_continuas
```

```python
#Criando uma função que preenche os vazios com o fill

def preenche_tabela(dados):
    features_continuas_colunas = dados.iloc[:, 13:-2].columns
    features_continuas = dados.groupby("PATIENT_VISIT_IDENTIFIER",
                                       as_index=False)[features_continuas_colunas].fillna(method='bfill').fillna(method='ffill')
    features_categoricas = dados.iloc[:, :13]
    saida = dados.iloc[:, -2:]
    dados_finais = pd.concat([features_categoricas, features_continuas, saida],
                             ignore_index=True,axis=1)
    dados_finais.columns = dados.columns
    return dados_finais
```

```python
dados_limpos = preenche_tabela(dados)
dados_limpos.head()
```

```python
dados_limpos = dados_limpos.dropna()
```

```python
#Precisamos retirar os pacientes que estao na WINDOW 0 ou 2 e ICU = 1, pois já
#estão na UTI

a_remover = dados_limpos.query("WINDOW=='0-2' and ICU==1")['PATIENT_VISIT_IDENTIFIER'].values
dados_limpos = dados_limpos.query("PATIENT_VISIT_IDENTIFIER not in @a_remover")
dados_limpos.describe()
```

```python
import numpy as np
```

```python
def prepare_window(rows):
    if(np.any(rows["ICU"])):
        rows.loc[rows["WINDOW"]=="0-2", "ICU"] = 1
    return rows.loc[rows["WINDOW"] == "0-2"]

dados_limpos = dados_limpos.groupby("PATIENT_VISIT_IDENTIFIER").apply(prepare_window)
dados_limpos.head()
```

```python
y = dados_limpos["ICU"]
x = dados_limpos[x_columns].drop(["ICU"], axis=1)
```

```python
np.random.seed(73246)
```

```python

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify = y)
```

```python
modelo = DummyClassifier()
modelo.fit(x_train, y_train)

y_prediction = modelo.predict(x_test)
accuracy_score(y_test, y_prediction)
```

```python
modelo = LogisticRegression()
modelo.fit(x_train, y_train)

y_prediction = modelo.predict(x_test)
accuracy_score(y_test, y_prediction)
```
